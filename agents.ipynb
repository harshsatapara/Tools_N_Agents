```python
from langchain_community.tools import DuckDuckGoSearchRun


search_tool = DuckDuckGoSearchRun()

search_result = search_tool.invoke('Top 5 latest news from India')

print(search_result)
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI()

```

```python
llm_result = llm.invoke('hi')
print(llm_result.content)
```

```python
from langchain.agents import create_react_agent,AgentExecutor
from langchain import hub

prompt = hub.pull("hwchase17/react")

agent = create_react_agent(llm=llm,tools=[search_tool],prompt=prompt)

agent_executor = AgentExecutor(agent=agent,tools=[search_tool],verbose=True)

result = agent_executor.invoke({"input" : "Which are the 3 way to reach Ahmedabad to Banglore"})

print(result)
```
